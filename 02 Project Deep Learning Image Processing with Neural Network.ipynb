```python
# converting RGB image to Grayscale through Open CV
import cv2
```

```python
pic = cv2.imread('/content/cat.jpg')
```

```python
type(pic)
```

```python
pic.shape
```

```python
grayscale = cv2.cvtColor(pic, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale)
```

```python
grayscale.shape
```
